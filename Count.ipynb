```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('freezer_image (1).jpg')
image2 = cv2.imread('freezer_image (2).jpg')
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray1, cmap='gray')
plt.imshow(gray2, cmap='gray')
```

```python
blur1 = cv2.GaussianBlur(gray1, (11, 11), 0)
blur2 = cv2.GaussianBlur(gray2, (11, 11), 0)
plt.imshow(blur1, cmap='gray')
plt.imshow(blur2, cmap='gray')
```

```python
canny1 = cv2.Canny(blur1, 30, 150, 3)
canny2 = cv2.Canny(blur2, 30, 150, 3)
plt.imshow(canny1, cmap='gray')
plt.imshow(canny2, cmap='gray')
```

```python
dilated1 = cv2.dilate(canny1, (1, 1), iterations=0)
plt.imshow(dilated1, cmap='gray')
dilated2 = cv2.dilate(canny2, (1, 1), iterations=0)
plt.imshow(dilated2, cmap='gray')
```

```python
(cnt1, hierarchy) = cv2.findContours(
    dilated1.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb1, cnt, -1, (0, 255, 0), 2)
  
plt.imshow(rgb1)

(cnt2, hierarchy) = cv2.findContours(
    dilated2.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb2, cnt, -1, (0, 255, 0), 2)
  
plt.imshow(rgb2)
```
